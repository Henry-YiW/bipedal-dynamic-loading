```python
from tensorboard.backend.event_processing.event_accumulator import EventAccumulator
import matplotlib.pyplot as plt

# Path to your .tfevents file
event_file = "pointfoot-legged-gym/logs/pointfoot_rough/Mar27_01-27-15_/events.out.tfevents.1743056835.henry-X570-AORUS-ULTRA.13681.0"

# Load event data
ea = EventAccumulator(event_file)
ea.Reload()

# Tags to plot
tags_colors = {
    'Train/mean_reward': 'tab:orange',
    'Loss/value_function': 'tab:orange'
}

# Plot each tag separately with assigned color
for tag, color in tags_colors.items():
    events = ea.Scalars(tag)
    steps = [e.step for e in events]
    values = [e.value for e in events]

    plt.figure()
    plt.plot(steps, values, color=color, label=tag)
    plt.xlabel("Step")
    plt.ylabel("Value")
    plt.title(f"{tag} over Steps")
    plt.legend()
    plt.grid(True)
    plt.show()
```
